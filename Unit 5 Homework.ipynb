```python
import numpy as np
import matplotlib.pyplot as plt
```

# Forward Euler

```python
def forward_euler(k, m, x0, v0, step, n):
  state_space = [[0, 1], [-k/m, 0]]
  solution = [x0, v0]
  steps = []
  solutions = []
  for i in range(n):
    solution = solution + (step*np.dot(state_space, solution))
    solutions.append(solution.tolist())
    t = (i+1)*step
    steps.append(t)
  return steps, solutions
```

```python
steps, solutions = forward_euler(k=2, m=0.5, x0=-1, v0=-2, step=0.1, n=5)
```

```python
steps, solutions
```

```python
steps_1, solutions_1 = forward_euler(k=2, m=0.5, x0=-1, v0=-2, step=0.1, n=200)
steps_2, solutions_2 = forward_euler(k=2, m=0.5, x0=-1, v0=-2, step=0.11, n=200)
```

```python
plt.scatter(np.array(solutions_1)[:,0], np.array(solutions_1)[:,1], label="0.1 Step")
plt.scatter(np.array(solutions_2)[:,0], np.array(solutions_2)[:,1], label="0.11 Step")
plt.legend()
```

# Backward Euler

```python
def backward_euler(k, m, x0, v0, step, n):
  state_space = [[0, 1], [-k/m, 0]]
  solution = [x0, v0]
  steps = []
  solutions = []
  for i in range(n):
    dummy_solution = solution + (step*np.dot(state_space, solution))
    solution = solution + (step*np.dot(state_space, dummy_solution))
    solutions.append(solution.tolist())
    t = (i+1)*step
    steps.append(t)
  return steps, solutions
```

```python
steps, solutions = backward_euler(k=2, m=0.5, x0=-1, v0=-2, step=0.1, n=5)
```

```python
steps, solutions
```

```python
steps_1, solutions_1 = backward_euler(k=2, m=0.5, x0=-1, v0=-2, step=0.1, n=200)
steps_2, solutions_2 = backward_euler(k=2, m=0.5, x0=-1, v0=-2, step=0.11, n=200)
```

```python
plt.scatter(np.array(solutions_1)[:,0], np.array(solutions_1)[:,1], label="0.1 Step")
plt.scatter(np.array(solutions_2)[:,0], np.array(solutions_2)[:,1], label="0.11 Step")
plt.legend()
```

```python
forward_steps, forward_solutions = forward_euler(k=2, m=0.5, x0=-1, v0=-2, step=0.1, n=50)
backward_steps, backward_solutions = backward_euler(k=2, m=0.5, x0=-1, v0=-2, step=0.1, n=50)
```
