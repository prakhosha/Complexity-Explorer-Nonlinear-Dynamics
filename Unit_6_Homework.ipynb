# Trapezoidal Method

```python
def trapezoid(k, m, x0, v0, step, n):
  state_space = [[0, 1], [-k/m, 0]]
  solution = [x0, v0]
  steps = []
  solutions = []
  for i in range(n):
    dummy_solution = solution + (step*np.dot(state_space, solution))
    solution = solution + step/2*(np.dot(state_space, solution)+np.dot(state_space, dummy_solution))
    solutions.append(solution.tolist())
    t = (i+1)*step
    steps.append(t)
  return steps, solutions
```
