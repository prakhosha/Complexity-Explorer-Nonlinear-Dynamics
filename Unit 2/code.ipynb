```python
# bifurcation diagram of logistic map function

# logistic map
def logistic_map(x0, r, n):
  x = x0
  x_list = []
  i_list = []
  r_list = []

  for i in range(n):
    x = r*x*(1-x)
    x_list.append(x)
    i_list.append(i)
    r_list.append(r)

  return i_list, x_list, r_list

# bifurcation
def bifurcation(x0, rmin, rmax, rstep, n, k):
  r_list = np.arange(rmin, rmax, rstep)
  r_bifur = []
  x_bifur = []

  for r in r_list:
    i_list, x_list, r_list = logistic_map(x0, r, n)
    r_bifur.append(r_list[k:])
    x_bifur.append(x_list[k:])

  return r_bifur, x_bifur
```

```python
# 5 to 1000

r, x = bifurcation(0.2, 2.4, 4, 0.01, 1000, 5)

plt.plot(r, x, ',k', alpha=0.25);
```

```python
# 500 to 1000

r, x = bifurcation(0.2, 2.4, 4, 0.01, 1000, 500)

plt.plot(r, x, ',k', alpha=0.25);
```
