```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# logistic map function

def log_map(x0, r, n):
  x = x0
  x_list = []
  i_list = []
  for i in range (n):
    x = r*x*(1-x)
    x_list.append(x)
    i_list.append(i)
  return i_list, x_list
```

```python
import matplotlib.pyplot as plt

n, x = log_map(0.2, 2.5, 3)

plt.scatter(n,x)
plt.xlabel("n")
plt.ylabel("x")
```
